## KNN分类原理

KNN方法：
1. 计算待分类样本与训练样本集中各个样本的距离；
2. 对距离进行排序；
3. 对前k个距离最小的样本进行类别统计（类似投票机制），样本最多的那一类即为待分类样本的类别

```python
from numpy import *
import operator

group = array([[1.0,1.1],[1.0,1.0],[0,0],[0,0.1]])
print group
```

```python
labels = ['A','A','B','B']
print labels
```

上面`group`为训练数据，`labels`为训练数据对应的类别标签

```python
inX= [0.3,0.3]
k = 3
dataSet = group
dataSetSize = dataSet.shape[0]    # instance number
diffMat = tile(inX, (dataSetSize, 1)) - dataSet
sqDiffMat = diffMat**2
distances = sqDiffMat.sum(axis = 1)    # 对维度求和
sortedDistIndicies = distances.argsort()
print sortedDistIndicies
print "max radius: ", distances[sortedDistIndicies[k-1]]
```

`inX`是某个测试的item，为了度量`inX`和训练数据集中各个item的相似性，采用欧式距离作为其相似性度量，分别计算与各个item的欧式距离，然后对计算的欧式距离进行排序，距离越小说明与item越相似，为了使其更加直观，下面对测试的item和训练集中的item进行可视化：

```python
import matplotlib.pyplot as plt
import pandas as pd

x = group[:,0]
y = group[:,1]
points = ['point:{0}'.format(i) for i in range(dataSetSize)]
df = pd.DataFrame(dict(x=x, y=y, label = labels, point = points))
groups = df.groupby('label')

# Plot
figure(figsize=(7,7))
ax=subplot(aspect='equal')
ax.margins(0.02) # Optional, just adds 5% padding to the autoscaling
for name, group in groups:
    ax.plot(group.x, group.y, marker='o', linestyle='', ms=12, label=name)
for point, x1, y1 in zip(points, x, y):
    ax.annotate(point, xy = (x1, y1), xytext = (-20, 20),textcoords = 'offset points', ha = 'right', va = 'bottom',bbox = dict(boxstyle = 'round,pad=0.5', fc = 'yellow', alpha = 0.5),arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3,rad=0'))
ax.legend(loc=2)
plt.plot(inX[0],inX[1],'ro',ms=12)
ax.annotate("A or B?", xy = (inX[0], inX[1]), xytext = (-20, 20),textcoords = 'offset points', ha = 'right', va = 'bottom',bbox = dict(boxstyle = 'round,pad=0.5', fc = 'yellow', alpha = 0.5),arrowprops = dict(arrowstyle = '->', connectionstyle = 'arc3,rad=0'))
circle1=plt.Circle((inX[0],inX[1]),distances[sortedDistIndicies[k-1]],color='r', fill=False)
plt.gcf().gca().add_artist(circle1)
plt.axis([-1,1.2,-1.2,1.2])
plt.show()
```

上图我们对训练集`group`中的item（蓝色为属于类别A的item，绿色为属于类别B的item）和测试item（红色所示），最近邻k设置为3，从上图可以看出，跟测试item最近的3个近邻是`point:1`、`point:2`和`point:3`，而且我们可以很直观的得到，测试item的类别为B。下面，将我们直观得出来的结果用代码实现完成：

```python
#对前k个距离最小的样本进行类别统计（类似投票机制），样本最多的那一类即为待分类样本的类别
classCount = {}
for i in range(k):
    voteIlabel = labels[sortedDistIndicies[i]]
    classCount[voteIlabel] = classCount.get(voteIlabel,0) + 1
sortedClassCount = sorted(classCount.iteritems(), key=operator.itemgetter(1),reverse=True)
print sortedClassCount[0][0]
```

## KNN分类练习

**任务**：使用KNN方法对约会对象进行分类

数据存放在`datingTestSet.txt`中，每一个样本数据占据一行，共四列，前三列为数值，每一列为一个特征，表示的意义如下所所示：
- 每年获得的飞行常客里程数
- 玩视频游戏所消耗的时间百分比
- 每周消费的冰淇淋公升数

第四列为类别标签，共三类，分别为：`largeDoses`,`smallDoses`,`didntLike`。共1000行，也即是1000个样本。

```python
from numpy import *

# 从文本文件中解析数据
fr = open("datingTestSet.txt")
arrayOlines = fr.readlines()
numberOfLines = len(arrayOlines)    # 得到样本数目：1000
returnMat = zeros((numberOfLines, 3))    # 用于保存特征
classLabelList = []
index = 0

classLabelsDict = {'largeDoses': 'green', 'smallDoses': 'red', 'didntLike': 'blue'}
Y = []

for line in arrayOlines:
    line  = line.strip()    # 去掉行尾\n的换行符
    listFromLine = line.split('\t')    # 以'\t'划分字符串
    returnMat[index, :] = listFromLine[0:3]
    classLabelList.append(listFromLine[3])
    Y.append(classLabelsDict[listFromLine[3]])
    index +=1

print returnMat
print classLabelList[0:20]
print Y[0:20]
```

上面我们将保存在`txt`文本文件中的数据解析成数值，并将其保存在一个名为`returnMat`的矩阵中，类别标记保存在`classLabelList`中，同时，为了后面画图所需，我们设置了一个`Y`的列表，用于为后面画的点进行着色。

从上面可视化的结果来看，**每年获取的飞行常客里程数**和**玩视频游戏所消耗的时间百分比**这两个特征对于该任务下的分类而言，是比较好的特征，能够更容易的区分数据点从属的类别，如果要做特征选择的话，我们可以使用该组特征。不过这里我们并不对特征进行选择，而是使用全部的3个特征。下面，在三维空间中，对其进行可视化：

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.margins(0.01) # Optional, just adds 5% padding to the autoscaling
ax.scatter(returnMat[:,0], returnMat[:,1], returnMat[:,2], s=30, c=Y)
ax.set_xlabel(u'每周消费的冰淇淋公升数',fontproperties=myfont)
ax.set_ylabel(u'玩视频游戏所消耗的时间百分比',fontproperties=myfont)
ax.set_zlabel(u'每周消费的冰淇淋公升数',fontproperties=myfont)
plt.show()
```

从三维视图俯视向下可以进一步的观察得到，增加的**每周消费的冰淇淋公升数**这个特征其实是冗余的，也就是这个特征是无关痛痒的，它的引入反而不利于分类。同时从上图可以发现，**每周消费的冰淇淋公升数**这一个特征其数字一般都非常的大，这样会使得在计算相似度（余弦距离）时，这个特征对计算的结果影响最大，而在这里，我们这三个特征是同等重要的，所以我们有必要采用对特征进行归一化的方式，消除这种影响。

归一化的方式有很多，这里采用如下方式将任意取值范围的特征值归一化到[0,1]之间：
newValue = (oldValue - min)/(max - min)
其中，min是**对应**特征的最小值，max是**对应**特征的最大值。注意这里的**对应**，一般在实际应用中，我们都会使用很多的特征，在对某一特征A归一化时，我们是在**训练集**中对特征A求的最大值和最小值，而不是在整个训练集上求最大值和最小值。

```python
# 特征归一化
dataSet = returnMat
minVals = dataSet.min(0)
maxVals = dataSet.max(0)
ranges = maxVals - minVals
normDataSet = zeros(shape(dataSet))
m = dataSet.shape[0]
normDataSet = dataSet - tile(minVals, (m,1))
normDataSet = normDataSet/tile(ranges, (m,1))
```

完成上述归一化后，我们在对其可视化一下，看看归一化随数据分布会有什么影响：

```python

import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

myfont = matplotlib.font_manager.FontProperties(fname='C:/Windows/Fonts/msyh.ttf')

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.margins(0.01) # Optional, just adds 5% padding to the autoscaling
ax.scatter(normDataSet[:,0], normDataSet[:,1], normDataSet[:,2], s=30, c=Y)
ax.set_xlabel(u'每周消费的冰淇淋公升数',fontproperties=myfont)
ax.set_ylabel(u'玩视频游戏所消耗的时间百分比',fontproperties=myfont)
ax.set_zlabel(u'每周消费的冰淇淋公升数',fontproperties=myfont)
plt.show()
```

从上图可以看出，归一化对数据分布不会造成任何的影响。经过归一化之后，各个特征是同等重要的。
